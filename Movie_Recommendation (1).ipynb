##**Aim : Building Movies Recommendation System using cosine similarity**

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
features=['keywords','cast','genres','director']
```

```python
for feature in features:
  df[feature]=df[feature].fillna('') # missing value will be  replaced by empty string
```

```python
def combined_features(row):
  return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
df["combined_features"]=df.apply(combined_features,axis=1)
```

```python
df["combined_features"].head()
```

```python
df.head()
```

```python
cv=CountVectorizer() # creating object for count vectorizer
count_matrix=cv.fit_transform(df['combined_features'])
print(count_matrix)
```

```python
cosine_sim=cosine_similarity(count_matrix)# to find the similarity between the vectors
```

```python
cosine_sim
```

```python
movies_user_likes="Tangled"
def get_index_from_title(title):
  return df[df.title==title]["index"].values[0]# return index of the movie
movie_index1=get_index_from_title(movies_user_likes)#extracts the index of the given movie
```

```python
movie_index1= int(movie_index1)
```

```python
 # Assuming cosine_sim is defined and contains the cosine similarity matrix
similar_movies = list(enumerate(cosine_sim[movie_index1,:]))
```

```python
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
def get_title(index):
  return df[df.index==index]["title"].values[0]
i=0
for movie in sorted_similar_movies:
  print(get_title(movie[1]))
  i=i+1
  if i>15:
    break
```
